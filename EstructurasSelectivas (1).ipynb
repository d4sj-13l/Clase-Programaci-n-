#Estructura Simple
si <condicion>entonces... fin si


```python
print("Ejemplo de Condición Simple")
print("Ingrese un número: ")
x=int(input())
if x<100:
  print("El numero es menor a 100")
```

#Estructura Selectiva Doble
si <condición> entonces
<acciones tipo 1:>
si no
<acciones tipo 2:>
fin si

```python
print("Ejemplo de Condición Doble")
print("Ingrese 2 números: ")
x=int(input("primer número: "))
y=int(input("segundo número: "))
if x>y:
   print("El menor es: ", y)
else:
   print("El menor es:", x)
```

#Estructura Selectiva Múltiple
si <selector> igual
   <valor1>: <operacion 1>
   <valor2>: <operacion 2>
   ...
   ...
   [<En otro caso>:<operacion n+1>]
  fin selector

```python
print("Ejemplo de condicional Múltiple")
switcher = {
   1: "Enero",
   2: "Febrero",
   3: "Marzo",
   4: "Abril",
   5: "Mayo",
   6: "Junio",
   7: "Julio",
   8: "Agosto",
   9: "Septiembre",
   10: "Octubre",
   11: "Noviembre",
   12: "Diciembre",
}
argument = int(input("Ingrese un número de mes a mostrar: "))
nombreMes = switcher.get(argument, "Mes inválido")
print(nombreMes)
input("Presione una tecla para salir")
print(nombreMes)
input("Presione una tecla para salir")

```
